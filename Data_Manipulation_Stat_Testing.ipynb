```python
#Question 13
sides = list(range(1,21))
sample = 10000
choice = random.choices(sides, k= sample)



```

```python
#Question 14. It appears to be a uniform distriution.
#Each number has the same chance of being picked.
plt.hist(choice)
plt.show
```

```python
#Question 15
#The CLT applies to our rolling die because these numbers have a mean of 0 and SD of 1

uniform_samples = [(random.uniform(0,1))for i in range(sample)]
normal_samples = [math.sqrt(-2 * math.log(u)) * math.cos(2 * math.pi * v) for u, v in zip(uniform_samples[::2], uniform_samples[1::2])]
```

```python
#Questiom 16
plt.hist(normal_samples)
plt.show()
```
